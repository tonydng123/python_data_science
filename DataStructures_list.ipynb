```python
# Lists
print("Lists")
node_1 = "custom object"

my_data_type = [1, 2, False, 4, "Tony", None, node_1, 5.0, 1]
print(my_data_type)
print(type(my_data_type))
print(my_data_type[4])

# Dictionary
print("Dictionary")
my_data_type = {1: "hello", 2: "Tony", 3: None, 4: node_1, 5: 5.0}
print(my_data_type)
print(type(my_data_type))
print(my_data_type[2])

# Set
print("Set")
my_data_type = {1, 2, False, 4, "Tony", None, node_1, 5.0, 1, 1}
print(my_data_type)
print(type(my_data_type))

# Tuples
print("Tuples")
my_data_type = (1, 2, False, 4, "Tony", None, node_1, 5.0, 1, 1)
print(my_data_type)
print(type(my_data_type))

# List
my_list = [5, 6, 7, 8, 35, 12, 14, 4, 10]
print(f"Memory location of my_list: {id(my_list)}") # same memory location
my_str_list =["Comp sci", "physics", "elec engr", "philosophy"]
print(f"Ints: {my_list}")
print(f"Strings: {my_str_list}")
print("Sorting...")
sorted_list = sorted(my_list)
print(sorted_list)
my_list.sort()
print(f"sort my list: {my_list}")
print(f"Memory location of sorted_list: {id(sorted_list)}")
print(f"Memory location of my_list: {id(my_list)}") # same memory location

print("Finding info...")
print("physics" in my_str_list)
print("None" in my_str_list)
print(35 in my_list)
print(34 in my_list)
print(my_str_list.index("physics"))
print(min(my_list))
print(min(my_str_list))
print(max(my_list))
print(max(my_str_list))

dir(my_list)
```
